## ‚úÖ ‡πÅ‡∏ö‡∏ö‡∏ù‡∏∂‡∏Å‡∏´‡∏±‡∏î 2.1 - SOLUTION (10 ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô)

**‡πÇ‡∏à‡∏ó‡∏¢‡πå:** ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô function `get_summary(text, max_length)`

````python
# ‚úÖ SOLUTION: ‡πÅ‡∏ö‡∏ö‡∏ù‡∏∂‡∏Å‡∏´‡∏±‡∏î 2.1

def get_summary(text, max_length=50):
    """
    ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏™‡∏£‡∏∏‡∏õ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°
    
    Args:
        text: ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ï‡πâ‡∏ô‡∏â‡∏ö‡∏±‡∏ö
        max_length: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏¢‡∏≤‡∏ß‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î (default: 50)
    
    Returns:
        str: ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡∏±‡∏î + "..." (‡∏ñ‡πâ‡∏≤‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô)
    """
    # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏¢‡∏≤‡∏ß
    if len(text) <= max_length:
        # ‡∏ñ‡πâ‡∏≤‡∏™‡∏±‡πâ‡∏ô‡∏Å‡∏ß‡πà‡∏≤‡∏´‡∏£‡∏∑‡∏≠‡πÄ‡∏ó‡πà‡∏≤‡∏Å‡∏±‡∏ö max_length ‚Üí return ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
        return text
    else:
        # ‡∏ñ‡πâ‡∏≤‡∏¢‡∏≤‡∏ß‡∏Å‡∏ß‡πà‡∏≤ ‚Üí ‡∏ï‡∏±‡∏î‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏ï‡∏¥‡∏° ...
        return text[:max_length] + "..."

# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö
print("Test 1:", get_summary("Hello", 10))           # "Hello" (‡∏™‡∏±‡πâ‡∏ô‡∏Å‡∏ß‡πà‡∏≤ 10)
print("Test 2:", get_summary("Hello World Today", 5)) # "Hello..." (‡∏¢‡∏≤‡∏ß‡∏Å‡∏ß‡πà‡∏≤ 5)
print("Test 3:", get_summary("ABCDE", 5))            # "ABCDE" (‡πÄ‡∏ó‡πà‡∏≤‡∏Å‡∏±‡∏ö 5)
````

## ‚úÖ ‡πÅ‡∏ö‡∏ö‡∏ù‡∏∂‡∏Å‡∏´‡∏±‡∏î 2.2 - SOLUTION (10 ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô)

**‡πÇ‡∏à‡∏ó‡∏¢‡πå:** ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô function `chunk_text(text, chunk_size)`

````python
# ‚úÖ SOLUTION: ‡πÅ‡∏ö‡∏ö‡∏ù‡∏∂‡∏Å‡∏´‡∏±‡∏î 2.2

def chunk_text(text, chunk_size=3):
    """
    ‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏õ‡πá‡∏ô chunks
    
    Args:
        text: ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ï‡πâ‡∏ô‡∏â‡∏ö‡∏±‡∏ö
        chunk_size: ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏≥‡∏ï‡πà‡∏≠ chunk
    
    Returns:
        list: ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£ chunks
    """
    # 1. ‡πÅ‡∏¢‡∏Å‡∏Ñ‡∏≥
    words = text.split()
    
    # 2. ‡∏™‡∏£‡πâ‡∏≤‡∏á list ‡πÄ‡∏Å‡πá‡∏ö chunks
    chunks = []
    
    # 3. ‡∏ß‡∏ô‡∏•‡∏π‡∏õ‡∏ó‡∏µ‡∏•‡∏∞ chunk_size ‡∏Ñ‡∏≥
    for i in range(0, len(words), chunk_size):
        # ‡∏î‡∏∂‡∏á‡∏Ñ‡∏≥‡∏ï‡∏±‡πâ‡∏á‡πÅ‡∏ï‡πà index i ‡∏ñ‡∏∂‡∏á i+chunk_size
        chunk_words = words[i:i + chunk_size]
        # ‡∏£‡∏ß‡∏°‡∏Ñ‡∏≥‡∏Å‡∏•‡∏±‡∏ö‡πÄ‡∏õ‡πá‡∏ô string
        chunk = " ".join(chunk_words)
        chunks.append(chunk)
    
    return chunks

# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö
sample = "one two three four five six seven"
result = chunk_text(sample, 2)
print("Chunks:", result)
# ['one two', 'three four', 'five six', 'seven']
````

# Part 3: Document Class - SOLUTIONS (30 ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô)

## ‚úÖ ‡πÅ‡∏ö‡∏ö‡∏ù‡∏∂‡∏Å‡∏´‡∏±‡∏î 3.1 - SOLUTION (30 ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô)

**‡πÇ‡∏à‡∏ó‡∏¢‡πå:** ‡∏™‡∏£‡πâ‡∏≤‡∏á Document class ‡∏ó‡∏µ‡πà‡∏™‡∏°‡∏ö‡∏π‡∏£‡∏ì‡πå

````python
# ‚úÖ SOLUTION: ‡πÅ‡∏ö‡∏ö‡∏ù‡∏∂‡∏Å‡∏´‡∏±‡∏î 3.1 - Document Class

class Document:
    """
    Class ‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏≠‡∏Å‡∏™‡∏≤‡∏£‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏£‡∏∞‡∏ö‡∏ö RAG
    """
    
    def __init__(self, title, content, source=None):
        """
        ‡∏™‡∏£‡πâ‡∏≤‡∏á Document ‡πÉ‡∏´‡∏°‡πà
        
        Args:
            title: ‡∏ä‡∏∑‡πà‡∏≠‡πÄ‡∏≠‡∏Å‡∏™‡∏≤‡∏£
            content: ‡πÄ‡∏ô‡∏∑‡πâ‡∏≠‡∏´‡∏≤
            source: ‡πÅ‡∏´‡∏•‡πà‡∏á‡∏ó‡∏µ‡πà‡∏°‡∏≤ (optional)
        """
        self.title = title
        self.content = content
        self.source = source
        
        # ‚úÖ ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì word_count ‡∏à‡∏≤‡∏Å content
        self.word_count = len(content.split())
    
    def get_summary(self, max_length=100):
        """
        ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏™‡∏£‡∏∏‡∏õ‡πÄ‡∏ô‡∏∑‡πâ‡∏≠‡∏´‡∏≤
        
        Returns:
            str: ‡πÄ‡∏ô‡∏∑‡πâ‡∏≠‡∏´‡∏≤‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡∏±‡∏î + "..." (‡∏ñ‡πâ‡∏≤‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô)
        """
        # ‚úÖ ‡πÉ‡∏ä‡πâ logic ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô function get_summary
        if len(self.content) <= max_length:
            return self.content
        return self.content[:max_length] + "..."
    
    def get_word_count(self):
        """
        ‡∏Ñ‡∏∑‡∏ô‡∏Ñ‡πà‡∏≤‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏≥
        
        Returns:
            int: ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏≥
        """
        # ‚úÖ return attribute word_count
        return self.word_count
    
    def to_dict(self):
        """
        ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô Dictionary
        
        Returns:
            dict: ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏≠‡∏Å‡∏™‡∏≤‡∏£
        """
        # ‚úÖ return dictionary ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏ó‡∏∏‡∏Å attribute
        return {
            "title": self.title,
            "content": self.content,
            "source": self.source,
            "word_count": self.word_count
        }
    
    def __repr__(self):
        """‡πÅ‡∏™‡∏î‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• object"""
        return f"Document(title='{self.title}', words={self.word_count})"
````

````python
# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö Document class

doc = Document(
    title="‡∏´‡∏±‡∏î‡πÄ‡∏¢‡∏≠‡∏£‡∏°‡∏±‡∏ô",
    content="‡∏´‡∏±‡∏î‡πÄ‡∏¢‡∏≠‡∏£‡∏°‡∏±‡∏ô‡πÄ‡∏õ‡πá‡∏ô‡πÇ‡∏£‡∏Ñ‡∏ï‡∏¥‡∏î‡∏ï‡πà‡∏≠‡∏ó‡∏µ‡πà‡πÄ‡∏Å‡∏¥‡∏î‡∏à‡∏≤‡∏Å‡πÑ‡∏ß‡∏£‡∏±‡∏™ ‡∏°‡∏µ‡∏≠‡∏≤‡∏Å‡∏≤‡∏£‡πÑ‡∏Ç‡πâ ‡∏ú‡∏∑‡πà‡∏ô ‡πÅ‡∏•‡∏∞‡∏ï‡πà‡∏≠‡∏°‡∏ô‡πâ‡∏≥‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡∏á‡πÇ‡∏ï",
    source="health.md"
)

print("1. Document:", doc)
print("2. Word count:", doc.get_word_count())
print("3. Summary:", doc.get_summary(30))
print("4. To dict:", doc.to_dict())
````

## ‚úÖ ‡πÅ‡∏ö‡∏ö‡∏ù‡∏∂‡∏Å‡∏´‡∏±‡∏î 4.1 - SOLUTION (15 ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô)

**‡πÇ‡∏à‡∏ó‡∏¢‡πå:** ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô function `process_documents(documents)`

````python
# ‚úÖ SOLUTION: ‡πÅ‡∏ö‡∏ö‡∏ù‡∏∂‡∏Å‡∏´‡∏±‡∏î 4.1

def process_documents(documents):
    """
    ‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡πÄ‡∏≠‡∏Å‡∏™‡∏≤‡∏£
    
    Args:
        documents: list ‡∏Ç‡∏≠‡∏á Document objects
    
    Returns:
        dict: ‡∏™‡∏£‡∏∏‡∏õ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏≠‡∏Å‡∏™‡∏≤‡∏£
    """
    # 1. ‡∏ô‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏≠‡∏Å‡∏™‡∏≤‡∏£
    total_documents = len(documents)
    
    # 2. ‡∏ß‡∏ô‡∏•‡∏π‡∏õ‡∏£‡∏ß‡∏° word_count ‡πÅ‡∏•‡∏∞‡πÄ‡∏Å‡πá‡∏ö title
    total_words = 0
    titles = []
    
    for doc in documents:
        total_words += doc.word_count  # ‡∏£‡∏ß‡∏°‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏≥
        titles.append(doc.title)        # ‡πÄ‡∏Å‡πá‡∏ö‡∏ä‡∏∑‡πà‡∏≠
    
    # 3. ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì average (‡∏£‡∏∞‡∏ß‡∏±‡∏á division by zero)
    if total_documents > 0:
        average_words = total_words / total_documents
    else:
        average_words = 0
    
    # 4. return dictionary
    return {
        "total_documents": total_documents,
        "total_words": total_words,
        "average_words": average_words,
        "titles": titles
    }

# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö
docs = [
    Document("Doc 1", "Hello World"),           # 2 ‡∏Ñ‡∏≥
    Document("Doc 2", "This is a longer document with more words"),  # 8 ‡∏Ñ‡∏≥
    Document("Doc 3", "Short text")             # 2 ‡∏Ñ‡∏≥
]

stats = process_documents(docs)
print("Statistics:", stats)
# {'total_documents': 3, 'total_words': 12, 'average_words': 4.0, 'titles': ['Doc 1', 'Doc 2', 'Doc 3']}
````

## ‚úÖ ‡πÅ‡∏ö‡∏ö‡∏ù‡∏∂‡∏Å‡∏´‡∏±‡∏î 4.2 - SOLUTION (15 ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô)

**‡πÇ‡∏à‡∏ó‡∏¢‡πå:** ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô function `search_documents(documents, query)`

````python
# ‚úÖ SOLUTION: ‡πÅ‡∏ö‡∏ö‡∏ù‡∏∂‡∏Å‡∏´‡∏±‡∏î 4.2

def search_documents(documents, query):
    """
    ‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤‡πÄ‡∏≠‡∏Å‡∏™‡∏≤‡∏£‡∏ó‡∏µ‡πà‡∏°‡∏µ query
    
    Args:
        documents: list ‡∏Ç‡∏≠‡∏á Document objects
        query: ‡∏Ñ‡∏≥‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤
    
    Returns:
        list: ‡πÄ‡∏≠‡∏Å‡∏™‡∏≤‡∏£‡∏ó‡∏µ‡πà‡∏ï‡∏£‡∏á‡∏Å‡∏±‡∏ô
    """
    # ‡πÅ‡∏õ‡∏•‡∏á query ‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡πá‡∏Å (case-insensitive)
    query_lower = query.lower()
    
    # ‡πÄ‡∏Å‡πá‡∏ö‡πÄ‡∏≠‡∏Å‡∏™‡∏≤‡∏£‡∏ó‡∏µ‡πà‡∏ï‡∏£‡∏á
    results = []
    
    for doc in documents:
        # ‡πÅ‡∏õ‡∏•‡∏á content ‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡πá‡∏Å‡πÅ‡∏•‡πâ‡∏ß‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö
        if query_lower in doc.content.lower():
            results.append(doc)
    
    return results

# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö
docs = [
    Document("‡∏´‡∏±‡∏î‡πÄ‡∏¢‡∏≠‡∏£‡∏°‡∏±‡∏ô", "‡∏´‡∏±‡∏î‡πÄ‡∏¢‡∏≠‡∏£‡∏°‡∏±‡∏ô‡πÄ‡∏õ‡πá‡∏ô‡πÇ‡∏£‡∏Ñ‡∏ï‡∏¥‡∏î‡∏ï‡πà‡∏≠ ‡∏°‡∏µ‡∏≠‡∏≤‡∏Å‡∏≤‡∏£‡πÑ‡∏Ç‡πâ"),
    Document("‡πÑ‡∏Ç‡πâ‡∏´‡∏ß‡∏±‡∏î", "‡πÑ‡∏Ç‡πâ‡∏´‡∏ß‡∏±‡∏î‡πÄ‡∏õ‡πá‡∏ô‡πÇ‡∏£‡∏Ñ‡∏ó‡∏≤‡∏á‡πÄ‡∏î‡∏¥‡∏ô‡∏´‡∏≤‡∏¢‡πÉ‡∏à ‡∏°‡∏µ‡∏≠‡∏≤‡∏Å‡∏≤‡∏£‡πÑ‡∏Ç‡πâ ‡πÑ‡∏≠"),
    Document("‡∏ï‡πâ‡∏≠‡∏Å‡∏£‡∏∞‡∏à‡∏Å", "‡∏ï‡πâ‡∏≠‡∏Å‡∏£‡∏∞‡∏à‡∏Å‡πÄ‡∏õ‡πá‡∏ô‡πÇ‡∏£‡∏Ñ‡∏ï‡∏≤ ‡∏û‡∏ö‡πÉ‡∏ô‡∏ú‡∏π‡πâ‡∏™‡∏π‡∏á‡∏≠‡∏≤‡∏¢‡∏∏")
]

results = search_documents(docs, "‡πÑ‡∏Ç‡πâ")
print(f"‡∏û‡∏ö {len(results)} ‡πÄ‡∏≠‡∏Å‡∏™‡∏≤‡∏£‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Ñ‡∏≥‡∏ß‡πà‡∏≤ '‡πÑ‡∏Ç‡πâ':")
for doc in results:
    print(f"  - {doc.title}")
# ‡∏û‡∏ö 2 ‡πÄ‡∏≠‡∏Å‡∏™‡∏≤‡∏£: ‡∏´‡∏±‡∏î‡πÄ‡∏¢‡∏≠‡∏£‡∏°‡∏±‡∏ô, ‡πÑ‡∏Ç‡πâ‡∏´‡∏ß‡∏±‡∏î
````
